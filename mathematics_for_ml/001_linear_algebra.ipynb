```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

Cop travels at 3km/min, the robber travels at 2.5km/min.
The robber has a head start of 5 mins.
At what time will the cop catch up to the robber.


```python
# T is the time, linear, n(1000) points
t = np.linspace(0, 40, 1000)
# Cop travelled distance 
distance_c = 3*(t-5) # as the robber has travelled for 5 mins already
# Robber travelled distance
distance_r = 2.5*t 

fix, ax = plt.subplots()
plt.title("Cop catching Robber")
plt.xlabel("Time (in minutes)")
plt.ylabel("Distance (in Km)")

ax.set_xlim([0, 45]) # Time is limited up to 40 mins, since he is caught before that
ax.set_ylim([0, 150]) # Distance can go up to hundred units
# plot the slopes and find intersection
ax.plot(t, distance_c, c='red')
ax.plot(t, distance_r, c='blue')
# Once you see results, draw intersection
plt.axvline(x=30, color="yellow", linestyle='--')
plt.axhline(y=75, color='yellow', linestyle='--')
```

On april 1, Mark1 design of solar panel starts generating power of 1Kj/day
On May 1 , Mark2 design of the solar panel starts generating power of 4kJ/day

```python
# From first iteration, its evident that they meet around 40, so limit x and y
t = np.linspace(0, 60, 1000) # timeperiod of 100 days
power_mark1 = 1*t
power_mark2 = 4*(t-30)
fix, ax = plt.subplots()
plt.title("Mark 1 vs Mark2 ")
plt.xlabel("Days")
plt.ylabel("KilloJoules of energy")
ax.set_xlim([0, 60])
ax.set_ylim([0, 80])
ax.plot(t, power_mark1, c="red")
ax.plot(t, power_mark2, c="blue")
plt.axvline(x=40, color="green", linestyle='--')
plt.axhline(y=40, color="green", linestyle='--')
```
