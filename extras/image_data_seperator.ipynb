```python
import os
import json
import sys
import shutil
```

```python
DATASET_BASE_DIR = '/home/arun/learn/tensorflow_certification/datasets'
BASE_DATASET = "food_101"
RAW_DATASET_DIR = "raw"
CUSTOM_DATASET = "mini_10"

ROOT_SRC_DIR = os.path.join(DATASET_BASE_DIR, BASE_DATASET, RAW_DATASET_DIR)
ROOT_DST_DIR = os.path.join(DATASET_BASE_DIR, BASE_DATASET, CUSTOM_DATASET)

print(ROOT_SRC_DIR)
print(ROOT_DST_DIR)
```

```python
classes_of_intrest = sorted([
"steak",
"omelette",
"pizza",
"sushi",
    "hamburger",
    "fried_rice",
    "nachos",
    "ramen",
    "hot_dog",
    "donuts",
    "ice_cream",
    "oysters",
    "onion_rings",
    "samosa",
    "spring_rolls",
    "tacos"
])
print(len(classes_of_intrest), classes_of_intrest)
```

```python
TRAIN_SIZE = 0.75
TEST_SIZE = 0.25
VALIDATE_SIZE = 0.0
```

```python
DATA_INFO = {}
with open(os.path.join(DATASET_BASE_DIR,BASE_DATASET,"meta", "train.json"), "r") as rfile:
    DATA_INFO["train"] = json.load(rfile)
with open(os.path.join(DATASET_BASE_DIR,BASE_DATASET,"meta", "test.json"), "r") as rfile:
    DATA_INFO["test"] = json.load(rfile)
```

```python
DATA_INFO["test"]["pizza"]
```

```python
for k in DATA_INFO:
    for food_class in DATA_INFO[k]:
        DATA_INFO[k][food_class] = [i.strip().split('/')[-1]+".jpg" for i in DATA_INFO[k][food_class]]
```

```python
for category in ["train", "test"]:
    base_dst_path = os.path.join(ROOT_DST_DIR, category)
    os.makedirs(base_dst_path, exist_ok=True)
    for food_class in classes_of_intrest:
        image_count = len(DATA_INFO[category][food_class])
        print(f"[{category}] the number of images in this category is : {image_count}")   
        base_src_path = os.path.join(ROOT_SRC_DIR, food_class)
        class_dst_path = os.path.join(base_dst_path, food_class)
        os.makedirs(class_dst_path, exist_ok=True)
        img_limit = round(image_count/10) if category == "train" else image_count
        print(f"[{category}] Image limit is {img_limit}")
        counter = 0
        for filename in os.listdir(base_src_path):
            # print(f"[{category}] the file is : {filename}")
            if counter >= img_limit:
                    break
            if filename in DATA_INFO[category][food_class]:
                src = os.path.join(base_src_path, filename)
                dst = os.path.join(class_dst_path, filename)
                # print(f"[{category}] Copying {src} to {dst}")
                shutil.copy(src, dst)
                counter += 1                
```
